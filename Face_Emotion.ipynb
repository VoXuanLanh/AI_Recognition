```python
from ultralytics import YOLO
from mtcnn import MTCNN
from matplotlib import pyplot as plt
import cv2

model = YOLO('/content/drive/MyDrive/Face recognition 2/runs/classify/train2/weights/best.pt')

img_path = '/content/drive/MyDrive/Face recognition 2/face/face1.jpg'
img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)

# D√πng MTCNN detect c√°c khu√¥n m·∫∑t
detector = MTCNN()
faces = detector.detect_faces(img)

plt.imshow(img)
ax = plt.gca()

for idx, face in enumerate(faces):
    x, y, width, height = face['box']
    x, y = max(0, x), max(0, y)

    face_crop = img[y:y+height, x:x+width]
    face_crop_resized = cv2.resize(face_crop, (224, 224))

    # L∆∞u t·∫°m ƒë·ªÉ predict
    temp_path = f'/content/face_tamthoi{idx}.jpg'
    cv2.imwrite(temp_path, cv2.cvtColor(face_crop, cv2.COLOR_RGB2BGR))

    results = model.predict(temp_path, verbose=False)
    probs = results[0].probs
    names = model.names

    predicted_class = names[probs.top1]
    confidence = probs.top1conf.item()*100

    rect = plt.Rectangle((x, y), width, height, fill=False, color='blue', linewidth=2)
    ax.add_patch(rect)

    label = f'{predicted_class} ({confidence:.1f}%)'
    plt.text(x, y, label, color='blue', fontsize=5, backgroundcolor='white')

    # V·∫Ω keypoints
    for key, value in face['keypoints'].items():
        dot = plt.Circle(value, radius=2, color='red')
        ax.add_patch(dot)

plt.axis('off')
plt.show()
!rm -rf /content/face_tamthoi*
```
